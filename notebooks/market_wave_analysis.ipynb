```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.fft
from tomlkit import date
df = pd.read_csv("googl_close.csv", parse_dates=['Date'])
df.head()
df
N = 19
```

```python
start = -np.pi
stop = np.pi+0.001
h = []
while (start < stop):
    h.append(1/(N*2+1) * np.sin(start*(N*2+1)/2) / np.sin(start/2))
    start += 0.001
H = np.array(h)
plt.plot(np.arange(-np.pi, np.pi+0.001, 0.001),
         20*np.log(np.abs(H))/np.log(10))
plt.show
plt.savefig("2-1.png")
```
