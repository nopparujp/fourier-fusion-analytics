```python
baboon, _ = librosa.load("baboon1.mp3")
bark, _ = librosa.load("bark.mp3")
bear, _ = librosa.load("bear-roar.mp3")
```

```python
def plot(signal,title):
    magnitude = np.abs(np.fft.fft(signal))
    frequency=np.linspace(0,22050,len(magnitude))
    plt.plot (frequency,magnitude)
    plt.title(title)
    plt.savefig(title+'.png')
    plt.show
    max_magnitude = np.abs(magnitude).max()
    return frequency[np.where(np.abs(magnitude)==max_magnitude)]
```

```python
plot(baboon,"baboon")
```

```python
plot(bark,"bark")
```

```python
plot(bear,"bear")
```
